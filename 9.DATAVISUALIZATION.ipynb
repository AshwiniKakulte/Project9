#### Bar Graph



```python
# Scatter plot with Age against Silver
plt.bar(data['Age'], data['Silver Medals'])
 
# Adding Title to the Plot
plt.title("Scatter Plot")
 
# Setting the X and Y labels
plt.xlabel('Age')
plt.ylabel('Silver Medals')
 
plt.show()
```



#### Scatter Plot



```python
# Scatter plot with Age against Silver Model
plt.scatter(data['Age'], data['Silver Medals'])
 
# Adding Title to the Plot
plt.title("Scatter Plot")
 
# Setting the X and Y labels
plt.xlabel('Age')
plt.ylabel('Silver Medals')
 
plt.show()
```

#### Bubble Chart



```python
sns.set_context("talk", font_scale=1.1)
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Gold Medals",
                y="Silver Medals",
                data=data)
# Put the legend out of the figure
plt.legend(bbox_to_anchor=(1.01, 1), borderaxespad=0)
plt.xlabel("Gold Medals")
plt.ylabel("Silver Medals")
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image
```

```python
text = open(r"C:\Users\Ashwini\OneDrive\Desktop\New Text Document (2).txt",
            mode='r', encoding='utf-8').read()
```

```python
mask = np.array(Image.open(r"C:\Users\Ashwini\OneDrive\Desktop\abc.png"))
```

```python
wc = WordCloud(stopwords = STOPWORDS,
               mask = mask, background_color = "white",
               max_words = 2000, max_font_size = 500,
               random_state = 42, width = mask.shape[1],
               height = mask.shape[0])
```

```python
wc.generate(text)
plt.imshow(wc, interpolation="None")
plt.axis('off')
plt.show()
```

#### Pie Chart


```python

df=data.groupby(['Country'])['Total Medals'].count().reset_index(name='Total Medals').sort_values(by= 'Total Medals', ascending= False).head(5)
```

```python
import plotly.express as px


```

```python
px.pie(df, values='Total Medals', names= 'Country' , title= 'Pie Chart', 
      
             color_discrete_map={'United States':'darkblue',
                                 'Russia':'cyan',
                                 'Germany':'magenta',
                              'Australia':'green','China':"yellow"})
```
